```python
import yfinance as yf
import pandas as pd
```

```python
i = 1
for _ in range(5):
    try:
        nifty50 = pd.read_html("https://en.wikipedia.org/wiki/NIFTY_50")[i].Symbol.to_list()
    except:
        i+=1
i = 1
for _ in range(5):
    try:
        niftynext50 = pd.read_html("https://en.wikipedia.org/wiki/NIFTY_Next_50")[i].Symbol.to_list()
    except:
        i+=1
nifty50tickers = [x + '.NS' for x in nifty50]
niftynext50tickers = [x + '.NS' for x in niftynext50]
tickers = nifty50tickers + niftynext50tickers
```

```python
# Define the start and end dates for which we want to get the data
start_date = '2010-01-01'
end_date = '2022-12-31'

# Download historical data
data = yf.download(tickers, start=start_date, end=end_date)
```

```python

# Select the adjusted closing prices
adj_close_prices = data['Adj Close']

# Calculate the correlation matrix
correlation_matrix = adj_close_prices.corr()

# Create a mask to disregard the diagonal and upper triangle of the correlation matrix
mask = pd.np.triu(pd.np.ones_like(correlation_matrix, dtype=bool))

# Zero out the upper triangle of the correlation matrix
tri_df = correlation_matrix.mask(mask)

# Stack the data and reset the index
pairs_to_drop = tri_df.stack().reset_index()

# Set column names
pairs_to_drop.columns = ['Ticker 1', 'Ticker 2', 'Correlation']

# Sort by correlation
pairs_to_drop.sort_values(by='Correlation', key=abs, ascending=False, inplace=True)

# Print pairs
pairs_to_drop.head(30)
```

```python
# Empty list to store the chosen pairs
chosen_pairs = []

# Empty set to store the tickers that have been used
used_tickers = set()

for _, row in pairs_to_drop.iterrows():
    if row['Ticker 1'] not in used_tickers and row['Ticker 2'] not in used_tickers:
        # Add the pair to the chosen pairs list
        chosen_pairs.append((row['Ticker 1'], row['Ticker 2']))
        
        # Add the tickers to the used tickers set
        used_tickers.update([row['Ticker 1'], row['Ticker 2']])
    
    # Stop if we have 30 pairs
    if len(chosen_pairs) == 30:
        break

# Print the chosen pairs
for pair in chosen_pairs:
    print(pair, correlation_matrix.loc[pair[0], pair[1]])
```

```python
from statsmodels.tsa.stattools import adfuller

# We'll use a significance level of 0.05
significance_level = 0.05

# Store the results
adf_results = []

# For each pair
for pair in chosen_pairs:
    # Calculate the spread
    try:
        spread = adj_close_prices[pair[0]] - adj_close_prices[pair[1]]
    except KeyError:
        print(f'Could not find {pair[0]} or {pair[1]} in the adjusted close prices.')
        continue
    
    try:
        # Run the ADF test on the spread
        result = adfuller(spread)
    except ValueError:
        print(f'ADF test failed for {pair}')
        continue
    
    # Get the p-value
    p_value = result[1]
    
    # Store the results
    adf_results.append({
        'Pair': pair,
        'ADF Statistic': result[0],
        'p-value': p_value,
        'Stationary': p_value < significance_level
    })

# Convert the results to a DataFrame
adf_results_df = pd.DataFrame(adf_results)

# Print the results
print(adf_results_df)
```
